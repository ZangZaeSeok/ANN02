# seed_chars='a' Test

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

dataset_path = '/workspace/ANN/days1/data/shakespeare_train.txt'
dset = dataset.Shakespeare(dataset_path)
best_model_path = '/workspace/ANN/days1/model/best_char_lstm_model.pth'
net = model.CharLSTM(len(dset.chars), 1024, len(dset.chars), dset.char_to_idx, dset.idx_to_char, n_layers=6).to(device)
net.load_state_dict(torch.load(best_model_path))
net.eval()

seed_chars = 'a'

for T in np.arange(0.5, 1.5, 0.1):
    print('-------------------------------------------------------------------')
    print(f'Temperature: {T}')
    print(generate.generate(net, seed_chars, temperature = T))
    print('-------------------------------------------------------------------')

print('-------------------------------------------------------------------')
print('Temperature: Down')
print(generate.generate(net, seed_chars, down_temperature=True, up_temperature=False))
print('-------------------------------------------------------------------')

print('-------------------------------------------------------------------')
print('Temperature: Up')
print(generate.generate(net, seed_chars, down_temperature=True, up_temperature=False))
print('-------------------------------------------------------------------')
```
